```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
#Loading the Dataset
```

```python
df=pd.read_csv("Iris.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.Species.value_counts()
```

```python
df.drop(["Id"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df["Species"]=df["Species"].str.replace("Iris-","")
```

```python
df.head()
```

```python
#EDA
```

```python
sns.pairplot(df, hue="Species", height=2.75)
```

```python
df.head()
```

```python
df["Species_Num"]=df["Species"]
df.head()
```

```python
df["Species_Num"]=df["Species_Num"].str.replace("setosa","1")
df["Species_Num"]=df["Species_Num"].str.replace("versicolor","2")
df["Species_Num"]=df["Species_Num"].str.replace("virginica","3")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Species_Num"]=pd.to_numeric(df["Species_Num"], downcast = "integer")
df.info()
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="Spectral")
plt.show()
```

```python
df.drop(["PetalLengthCm", "Species_Num"], axis=1, inplace= True)
```

```python
df.head()
```

```python
#Decision Tree Modelling
```

```python
X=df.drop(["Species"],axis=1)
Y=df["Species"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train, Y_test = train_test_split(X,Y,test_size =0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier()
```

```python
dt_model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
Y_pred = dt_model.predict(X_test)
print("Accuracy Score:", accuracy_score(Y_test, Y_pred))
```

```python
from sklearn.model_selection import cross_val_score #K-Fold cross validation
```

```python
cv_scores = cross_val_score(estimator=dt_model, X=X_train, y=Y_train, cv=10)
cv_scores.mean()
```

```python
cv_scores
```

```python
print(confusion_matrix(Y_test,Y_pred))
```

```python
cnf_matrix = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cnf_matrix, annot=True, cmap="YlGnBu")
plt.ylabel("Actual label")
plt.xlabel("Predicted label")
```

```python
print(classification_report(Y_test, Y_pred))
```
